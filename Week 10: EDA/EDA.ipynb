### Persistency by gender

```python

# Group the data by Gender and Persistency_Flag and count the occurrences of each combination
counts = df.groupby(['Gender', 'Persistency_Flag']).size()

# Calculate the percentages of each combination
percentages = counts.groupby(level=0).apply(lambda x: x / float(x.sum()))

# Plot the results as a bar chart
ax = percentages.plot(kind='bar')
ax.set_ylabel('Percentage')
ax.set_title('Percentage of Persistent and Non-persistent Patients by Gender')
plt.show()
```

### For the following plots, '0' = Yes and '1' = No

```python
def plot_cols(x, df):
    for col in x:
        counts = df[col].value_counts()
        total = counts.sum()
        percentages = counts / total * 100
        fig, ax = plt.subplots(figsize=(10, 4))
        ax.bar(counts.index.astype(str), counts.values)
        ax.set_yticks(range(0, counts.max()+1, int(counts.max()/10)))
        ax.set_ylabel('count')
        ax.set_xlabel(col)
        ax.set_title(f'Counts for {col}')
        for i, v in enumerate(counts.values):
            ax.text(i, v, f'{v} ({percentages.iloc[i]:.2f}%)', ha='center', va='bottom')
        plt.xticks(rotation=90)
        plt.show()
plot_cols(cols, df)

```

```python
obj_cols = df.select_dtypes(include=['object']).columns

def plot_objcols(x, df):
    for col in x:
        counts = df[col].value_counts()
        percentage = counts / counts.sum() * 100
        fig, ax = plt.subplots(figsize=(10, 4))
        graph = counts.plot(kind='barh')
        plt.xticks(rotation=90)
        plt.title('Counts for ' + col)
        plt.xlabel('count')
        # plotting the graph with percentages
        for i, p in enumerate(graph.patches):
            pct = f'{percentage[i]:.2f}%'
            x1, y1 = p.get_width(), p.get_y() + p.get_height() / 2
            ax.annotate(pct, (x1, y1))
        plt.show()

plot_objcols(obj_cols, df)
```

```python
# Replace 1 with Fracture and 0 with No Fracture
df['Frag_Frac_During_Rx'] = df['Frag_Frac_During_Rx'].replace({1: 'Fracture', 0: 'No Fracture'})

# Group the data by the two variables and count the number of occurrences
counts = df.groupby(['Frag_Frac_During_Rx', 'Persistency_Flag']).size()

# Calculate the percentages of each combination
percentages = counts.groupby(level=0).apply(lambda x: x / float(x.sum()))

# Plot the results as a bar chart
ax = percentages.plot(kind='bar')
ax.set_ylabel('Percentage')
ax.set_title('Percentage of Persistent and Non-persistent Patients by Fragment Fracture During Rx')
plt.show()
```

```python
grouped = df.groupby(['Change_T_Score', 'Persistency_Flag']).size().unstack()

# Calculate the percentage of each category within each group
grouped_pct = grouped.apply(lambda x: 100 * x / float(x.sum()), axis=1)

# Plot the grouped data using a bar chart
ax = grouped_pct.plot(kind='bar', stacked=True)
ax.set_xlabel('Change T Score')
ax.set_ylabel('Percentage')
plt.ylim([0, 100]) # Set y-axis limit to 0-100%

# Add percentage labels to the bars
for i in ax.containers:
    ax.bar_label(i, label_type='edge', labels=[f'{h:.1f}%' for h in i.datavalues], label_padding=4)

plt.show()
```
